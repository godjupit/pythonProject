```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2*np.pi, 2*np.pi,50)
y = np.sin(x)
plt.figure(figsize=(10,8))
plt.scatter(x = x,y = y) #散点图
plt.title('y = sin(x)')
plt.xlabel('x')  # x title
plt.ylabel('y')
plt.show()
```
